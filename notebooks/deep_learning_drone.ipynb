```python
from ultralytics import YOLO
import shutil

DATA_YAML = "data/data.yaml"
```

```python
model = YOLO("yolov5s.yaml")

results = model.train(
    data=DATA_YAML,
    epochs=50,
    imgsz=640,
    batch=16,
    pretrained=False,
    project="runs",
    name="yolov5-scratch",
)

val_metrics = model.val(split="val")
test_metrics = model.val(split="test")
print(f"Val  mAP@50: {val_metrics.box.map50:.4f}")
print(f"Test mAP@50: {test_metrics.box.map50:.4f}")
```

```python
model = YOLO("yolov5s.pt")

results = model.train(
    data=DATA_YAML,
    epochs=50,
    imgsz=640,
    batch=16,
    project="runs",
    name="yolov5-pretrained",
)

val_metrics = model.val(split="val")
test_metrics = model.val(split="test")
print(f"Val  mAP@50: {val_metrics.box.map50:.4f}")
print(f"Test mAP@50: {test_metrics.box.map50:.4f}")
```
