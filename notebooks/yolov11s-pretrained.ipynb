```python
!pip install ultralytics
from ultralytics import YOLO
import shutil
DATA_YAML = "data/data.yaml"
```

```python
model = YOLO("yolo11s.pt")
model.train(
  data=DATA_YAML,
  epochs=250,
  imgsz=640,
  batch=16,
  lr0=0.01,
  cos_lr=True,
  patience=40,
  project="/home/sagemaker-user/runs",
  name="yolo11s-v2",
)


shutil.copy("/home/sagemaker-user/runs/yolo11s-v2/weights/best.pt", "models/yolo11s-v2.pt")
model = YOLO("models/yolo11s-v2.pt")
val_metrics = model.val(split="val")
test_metrics = model.val(split="test")
print(f"Val  mAP@50: {val_metrics.box.map50:.4f}")
print(f"Test mAP@50: {test_metrics.box.map50:.4f}")
```
